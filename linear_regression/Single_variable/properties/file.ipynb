```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
csv_data = pd.read_csv("homeprices.csv")
csv_data
```

```python
rg = linear_model.LinearRegression()
rg.fit(csv_data[['area']], csv_data.price)
```

```python
%matplotlib inline
plt.xlabel("area (sq ft)")
plt.ylabel("price (USD)")
plt.scatter(csv_data.area, csv_data.price, color = "red", marker = "+")
plt.plot(csv_data.area, rg.predict(csv_data[['area']]), color = "blue")
```
